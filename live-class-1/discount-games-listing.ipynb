```python
import requests
import pandas as pd
```

```python
resp = requests.get("https://www.cheapshark.com/api/1.0/deals?upperPrice=30&pageSize=1000")
resp.status_code
```

```python
raw_data = resp.json()
raw_data
```

```python
game_data = pd.DataFrame.from_dict(raw_data)
game_data
```

```python
game_data.info()
```

```python
get_shape = game_data.shape

print(f"Data Game Listing memiliki shape {get_shape}")
```

```python
SELECTED_COLS = ["title","storeID","gameID","salePrice","normalPrice","steamRatingText","steamRatingCount","releaseDate","lastChange"]
selected_game_data= game_data[SELECTED_COLS]

selected_game_data
```

```python
selected_game_data.loc[1:40]
```

```python
RENAME_COLS = {
    "storeID" : "store_id",
    "gameID" : "game_id",
    "salePrice" : "sale_price",
    "normalPrice" : "normal_price",
    "steamRatingText" : "steam_rating_text",
    "steamRatingCount" : "steam_rating_count",
    "releaseDate" : "release_date",
    "lastChange": "last_change"
}
```

```python
rename_game_data = selected_game_data.rename(columns=RENAME_COLS)
rename_game_data
```

```python
pd.to_datetime(rename_game_data["release_date"], unit = 's')
rename_game_data["release_date"] = pd.to_datetime(rename_game_data["release_date"], unit = 's')

rename_game_data["release_date"]
```

```python
rename_game_data["last_change"] = pd.to_datetime(rename_game_data["last_change"], unit = 's')

rename_game_data["last_change"]
```

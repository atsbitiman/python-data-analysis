```python
import pandas as pd
import requests

# get API dari PokeAPI
resp = requests.get("https://pokeapi.co/api/v2/pokemon?limit=500")
# cek status code dari API, 200 artinya mengembalikan response
resp.status_code
```

```python
# output API
resp.text
```

```python
# ubah API data menjadi bentuk JSON
raw_data = resp.json()
raw_data
```

```python
pokemon_data = pd.DataFrame(raw_data['results'])
pokemon_data
```

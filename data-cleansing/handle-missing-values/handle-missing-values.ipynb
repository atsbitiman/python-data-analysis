```python
import pandas as pd
```

```python
property_data = pd.read_csv('property_data.csv')
property_data
```

```python
property_data.dtypes
```

```python
# meghitung nilai mean pada kolom SQ_FT
med_SQ = property_data["SQ_FT"].median()
print(med_SQ)
```

```python
# Menginput NaN pada kolom SQ_FT dengan nilai median
property_data["SQ_FT"] = property_data["SQ_FT"].fillna(med_SQ)
property_data
```

```python
# Mencari nilai modus dari kolom NUM_BEDROOMS dan NUM_BATH
mod_NUM_BEDROOMS = property_data["NUM_BEDROOMS"].mode()[0]
mod_NUM_BATH = property_data["NUM_BATH"].mode()[0]

# Menginput NaN pada kolom NUM_BEDROOMS dan NUM_BATH dengan nilai modusnya masing-masing
property_data["NUM_BEDROOMS"] = property_data["NUM_BEDROOMS"].fillna(mod_NUM_BEDROOMS)
property_data["NUM_BATH"] = property_data["NUM_BATH"].fillna(mod_NUM_BATH)
property_data
```

```python
# Mencari nilai modus dari kolom OWN_OCCUPIED
mod_OWN_OCCUPIED = property_data["OWN_OCCUPIED"].mode()[0]

# Menginput Modus pada kolom OWN_OCCUPIED
property_data["OWN_OCCUPIED"] = property_data["OWN_OCCUPIED"].fillna(mod_OWN_OCCUPIED)
property_data
```

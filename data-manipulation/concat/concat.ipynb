```python
import pandas as pd
```

```python
df1 = {
    "pet" : ["Dog","Cat"],
    "color" : ["Black","White"]
}

df1 = pd.DataFrame(df1)
df1
```

```python
df2 = {
    "pet" : ["fish","turtle"],
    "color" : ["Yellow", "Black"]
}

df2 = pd.DataFrame(df2)
df2
```

```python
df_concat = pd.concat([df1,df2])
df_concat
```

```python
df_concat = pd.concat([df1,df2], ignore_index=True)
df_concat
```

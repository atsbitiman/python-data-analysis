```python
import pandas as pd
```

```python
df1 = {
    "pet" : ["Dog","Cat"],
    "color" : ["Black","White"]
}

df1 = pd.DataFrame(df1)
df1
```

```python
df2 = {
    "pet" : ["fish","turtle"],
    "color" : ["Yellow", "Black"]
}

df2 = pd.DataFrame(df2)
df2
```

```python
df_concat = pd.concat([df1,df2])
df_concat
```

```python
df_concat = pd.concat([df1,df2], ignore_index=True)
df_concat
```

```python
df_concat = pd.concat([df1,df2], keys=['darat','laut'])
df_concat
```

```python
# Memilih data dengan keys 'darat'
df_concat.loc['darat']
```

```python
df1 = {
    "pet" : ["Dog","Cat"],
    "name" : ["Bleki","Alexa"]
}

df2 = {
    "pet" : ["Dog","Cat","Fish"],
    "color" : ["Black","White","Yellow"]
}

df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
```

```python
df1
```

```python
df2
```

```python
df_concat = pd.concat([df1,df2])
df_concat
```

```python
# Concat kolom - Menggabungkan kolom
# Buat dataframe baru untuk menyimpan hasil concat

df_concat = pd.concat([df1,df2], axis=1)
df_concat
```

```python
# Concat kolom dengan tipe inner join - Menggabungkan kolom
# Buat dataframe baru untuk menyimpan hasil concat
df_concat = pd.concat([df1,df2], axis = 1, join="outer")
df_concat
```

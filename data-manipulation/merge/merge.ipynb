```python
import pandas as pd
```

```python
opinion = pd.read_csv("opinion_climate_change.csv")
opinion
```

```python
opinion_author = pd.read_csv("opinion_climate_change_author.csv")
opinion_author
```

```python
opinion_merge = opinion.merge(opinion_author, left_on="comment_id", right_on="id_comment")
opinion_merge
```

```python
opinion_merge = opinion.merge(opinion_author, left_index=True, right_index=True)
opinion_merge
```

```python
# Buat dataframe baru untuk menyimpan hasil merge
opinion_merge = opinion.merge(opinion_author,
                              left_on = 'comment_id',
                              right_on = 'id_comment',
                              how = 'left')

# Menampilkan 5 data teratas
opinion_merge.head()
```

```python
# Buat dataframe baru untuk menyimpan hasil merge
opinion_merge = opinion.merge(opinion_author,
                              left_on = 'comment_id',
                              right_on = 'id_comment',
                              how = 'right')

# Menampilkan 5 data teratas
opinion_merge.head()
```

```python
# Buat dataframe baru untuk menyimpan hasil merge
opinion_merge = opinion.merge(opinion_author,
                              left_on = 'comment_id',
                              right_on = 'id_comment',
                              how = 'inner')

# Menampilkan 5 data teratas
opinion_merge.head()
```

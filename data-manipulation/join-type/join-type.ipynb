```python
import pandas as pd
```

```python
df1 = {
    "pet" : ["Dog","Cat","Rabbit","Fish","Turtle"]
}

df1 = pd.DataFrame(df1)
df1
```

```python
df2 = {
    "color" : ["Black","Blue","Red","Green","Yellow","White"]
}

df2 = pd.DataFrame(df2)
df2
```

```python
df_join = df1.join(df2)
df_join
```

```python
df_join = df1.join(df2, how="right")
df_join
```

```python
df_join = df1.join(df2, how="outer")
df_join
```

```python
df_join = df1.join(df2, how="inner")
df_join
```

```python
climate_change = pd.read_csv("reddit_opinion_climate_change.csv")
climate_change
```

```python
climate_change_author = pd.read_csv("reddit_opinion_climate_change_author.csv")
climate_change_author
```

```python
df_join = climate_change.join(climate_change_author, how="left")
df_join
```

```python
df_join = climate_change.join(climate_change_author, how="right")
df_join
```

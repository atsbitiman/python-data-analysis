```python
import pandas as pd
```

```python
superstore_df = pd.read_csv("superstore.csv")
superstore_df.head()
```

```python
# Select data
superstore_df = superstore_df[['Category', 'Order.Date', 'Profit']]

# Menampilkan 5 data teratas
superstore_df.head()
```

```python
superstore_df.info()
```

```python
# Casting data
superstore_df['Order.Date'] = pd.to_datetime(superstore_df['Order.Date'])

# Menampilkan 5 data teratas
superstore_df.head()
```

```python
superstore_df.info()
```

```python
# Membuat kolom "year_order" untuk menyimpan hasil ekstraksi tahun dari kolom "Order.Date"
superstore_df['year_order'] = superstore_df['Order.Date'].dt.year

# Membuat kolom "month_order" untuk menyimpan hasil ekstraksi bulan dari kolom "Order.Date"
superstore_df['month_order'] = superstore_df['Order.Date'].dt.month

# Membuat kolom "day_order" untuk menyimpan hasil ekstraksi tanggal dari kolom "Order.Date"
superstore_df['day_order'] = superstore_df['Order.Date'].dt.day

# Menampilkan 5 data teratas
superstore_df.head()
```

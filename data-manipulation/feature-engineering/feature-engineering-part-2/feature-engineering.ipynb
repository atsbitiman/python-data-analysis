```python
import pandas as pd
```

```python
superstore_df = pd.read_csv("superstore.csv")

superstore_df = superstore_df[["Category","Order.Date","Profit"]]
superstore_df
```

```python
superstore_df.info()
```

```python
# Casting data
superstore_df['Category'] = superstore_df['Category'].astype('category')

# Menampilkan 5 data teratas
superstore_df.head()
```

```python
# Menampilkan info tipe data dari data superstore
superstore_df.dtypes
```

```python
# generate label dari kolom `Category` dan disimpan kedalam kolom baru `category_label`
superstore_df['category_label'] = superstore_df['Category'].cat.codes

# Menampilkan 5 data teratas
superstore_df.head()
```

```python
# Menampilkan baris pertama hasil grouping berdasarkan category
superstore_df.groupby('Category').first()
```

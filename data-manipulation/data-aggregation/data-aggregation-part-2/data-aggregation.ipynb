```python
import pandas as pd
```

```python
cardiovascular_df = pd.read_csv("Cardiovascular_Disease.csv")
cardiovascular_df.head()
```

```python
# Mencari nilai minimal dan maksimal dengan axis = 0
cardiovascular_df.agg(func= ['min','max'], axis = 0)
```

```python
# Mencari nilai minimal dan maksimal dengan axis = 1
cardiovascular_df.agg(func=['min','max'], axis=1)
```

```python
# Melakukan aggregasi pada kolom dan fungsi aggregasi yang berbeda
cardiovascular_df.agg({'age' : ['max'], 'serumcholestrol' : ['max', 'mean']})
```

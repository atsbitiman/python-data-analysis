```python
import pandas as pd
```

*Menghitung Rata - Rata*

```python
superstore_df = pd.read_csv("superstore.csv")
superstore_df
```

```python
# Menampilkan info dari data superstore
superstore_df.info()
```

```python
# Mencari rata-rata pada kolom Sales
sales_mean = superstore_df.Sales.mean()
print(f"Nilai rata-rata sales : {round(sales_mean,2)}")
```

*Mencari Standard Deviasi*

```python
sales_std = superstore_df.Sales.std()
print(f"Nilai standard deviasi sales : {round(sales_std,2)}")
```

```python
sales_median = superstore_df.Sales.median()
print(f"Nilai median Sales : {round(sales_median,2)}")
```

```python
sales_max = superstore_df.Sales.max()
print(f"Nilai terbesar Sales : {round(sales_max,2)}")
```

```python
sales_min = superstore_df.Sales.min()
print(f"Nilai terkecil dari Sales : {round(sales_min,2)}")
```

```python
sales_count = superstore_df.Sales.count()
print(f"Jumlah baris dari Sales : {round(sales_count,2)}")
```

```python
sales_mode = superstore_df.Sales.mode()
print(f"Jumlah mode dari Sales : {(sales_mode)}")
```

```python
Category_nunique = superstore_df.Category.nunique()
print(f"Jumlah nilai unique dalam kolom Category : {Category_nunique}")
```

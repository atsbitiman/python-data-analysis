```python
import pandas as pd
```

```python
superstore = pd.read_csv("superstore.csv")
superstore
```

```python
# Buat object untuk menyimpan hasil grouping
superstore_category = superstore.groupby(by="Category")
superstore_category
```

```python
superstore_category.first()
```

```python
superstore_category.size()
```

```python
# Melihat 5 data dari setiap kategori
for category, dataframe in superstore_category : 
    print(category)
    print("-------------")
    print(dataframe.head())
    print("")
    print("")
```

```python
# Melihat hasil grouping untuk Category = Technology
superstore_category.get_group('Technology').head()
```

```python
# Buat object untuk menyimpan hasil grouping
# Grouping berdasarkan kolom Category dan Market
superstore_category_market = superstore.groupby(by = ['Category', 'Market'])
superstore_category_market.first()
```

```python
# Melihat jumlah data untuk setiap category dan market
superstore_category_market.size()
```

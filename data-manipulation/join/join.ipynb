```python
import pandas as pd
```

```python
product_data = pd.read_csv("product.csv")
product_data
```

```python
order_data = pd.read_csv("order.csv")
order_data
```

```python
# Left Join
data_join = product_data.join(order_data, lsuffix = "_left")
data_join
```

```python
data_join = product_data.join(order_data, rsuffix = "_right")
data_join
```

```python
# Buat dataframe baru yang akan menyimpan data hasil join
# Menambahkan parameter lsuffix dan rsuffix
data_join = product_data.join(order_data, lsuffix = "_left", rsuffix = "_right")

# Menampilkan 5 data teratas
data_join.head()
```

```python
import pandas as pd
```

```python
house_pricing = pd.read_csv("house-pricing.csv")
house_pricing.head()
```

```python
house_spesification = pd.read_csv("house-spesifications.csv")
house_spesification.head()
```

```python
house_spesification.info()
```

```python
house_spesification.set_index("house_id", inplace=True)
house_spesification.head()
```

```python
house_pricing.set_index("house_id", inplace=True)
house_pricing.head()
```

```python
house_joined = house_spesification.join(house_pricing,how='inner')
house_joined.head()
```

```python
get_lower_price = house_joined["Price"].min()
print(f"Harga Rumah terendah adalah {get_lower_price}")
```

```python
get_higher_price = house_joined["Price"].max()
print(f"Harga Rumah tertinggi adalah {get_higher_price}")
```

```python
get_mean_price = house_joined["Price"].mean()
print(f"Harga Rata - Rata rumah adalah {get_mean_price}")
```

```python
house_joined[house_joined["Price"] == get_lower_price]
```

```python
house_joined[house_joined["Price"] == get_higher_price]
```

```python
house_joined[house_joined["Price"] <= get_mean_price]
```

```python
house_joined[house_joined["Price"] >= get_mean_price]
```

```python
BINNING_SQUARE_FEET = [1000, 1500, 2000, 2500, 3000]

LABEL_BINNING = ["Small Homes","Medium Homes","Midsize Homes","Large Homes"]

house_joined["house_category"] = pd.cut(house_joined["SquareFeet"], bins = BINNING_SQUARE_FEET, labels=LABEL_BINNING)
house_joined
```

```python
house_category = house_joined.groupby(["house_category"]).count()
```

```python
house_joined[house_joined["house_category"] == "Small Homes"].head()
```

```python
house_joined[house_joined["house_category"] == "Medium Homes"].head()
```

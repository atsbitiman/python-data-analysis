```python
import pandas as pd
```

```python
# Read Data
spotify_tracks = pd.read_csv("spotify-track.csv")

# Tampilkan 5 data teratas
spotify_tracks.head()
```

```python
# Tampilkan info
spotify_tracks.info()
```

```python
# Read data
spotify_albums = pd.read_csv("spotify-album.csv")

# Tampilkan 5 data teratas
spotify_albums.head()
```

```python
# Tampilkan info
spotify_albums.info()
```

```python
# Lakukan merge
# Buat dataframe baru untuk menyimpan hasil merge
spotify_merged = spotify_tracks.merge(spotify_albums,
                                      left_on = "track_id",
                                      right_on = "id_track")

# Tampilkan 5 data teratas
spotify_merged.head()
```

```python
# Hapus data yang duplikat pada track_id, jika ada
spotify_merged.drop_duplicates(subset = "track_id",
                               inplace = True,
                               ignore_index = True)

# Tampilkan 5 data teratas
spotify_merged.head()
```

```python
# Drop kolom "id_track"
spotify_merged.drop(columns = 'id_track',
                    inplace = True)

# Tampilkan 5 data teratas
spotify_merged.head()
```

```python
# Merubah kolom 'track_album_release_date' menjadi datetime
spotify_merged['track_album_release_date'] = pd.to_datetime(spotify_merged['track_album_release_date'])

# Menampilkan 5 data teratas
spotify_merged.head()
```

```python
# Menampilkan tipe data dari masing-masing kolom
spotify_merged.dtypes
```

```python
# Ekstrak data tahun dari kolom 'track_album_release_date'
spotify_merged["year_album"] = spotify_merged["track_album_release_date"].dt.year

# Ekstrak data bulan dari kolom 'track_album_release_date'
spotify_merged["month_album"] = spotify_merged["track_album_release_date"].dt.month

# Ekstrak data tanggal dari kolom 'track_album_release_date'
spotify_merged["day_album"] = spotify_merged["track_album_release_date"].dt.day

# Hapus kolom 'track_album_release_date'
spotify_merged.drop(columns = "track_album_release_date", inplace = True)

# Menampilkan 5 data teratas
spotify_merged.head()
```

```python
# Mendapakan data dengan track_popularity paling tinggi
popularity_max = spotify_merged[spotify_merged['track_popularity'] == spotify_merged['track_popularity'].max()]

# Print tahun dengan track_popularity paling tinggi
print(f"Track popularity paling tinggi terdapat pada tahun : {popularity_max.year_album.values[0]}")

# Menampilkan data dengan track_popularity paling tinggi
spotify_merged[spotify_merged["track_popularity"] == spotify_merged['track_popularity'].max()]
```

```python
# Menggunakan groupby untuk mengelompokkan data berdasarkan kolom 'year_album'
spotify_year_group = spotify_merged.groupby('year_album')
```

```python
# Menggunakan fungsi idxmax() untuk mendapatkan indeks track_popularity maksimum untuk setiap kelompok year_album
indeks_max = spotify_year_group['track_popularity'].idxmax()

# Memfilter DataFrame asli berdasarkan indeks maksimum yang telah diketahui sebelumnya
spotify_filtered = spotify_merged.loc[indeks_max]

# Mengurutkan DataFrame hasil filter berdasarkan kolom 'year_album' secara descending
spotify_filtered.sort_values('year_album',
                             ascending=False,
                             inplace=True)
    
# Menampilkan 5 baris pertama dari DataFrame hasil filter
spotify_filtered.head()
```

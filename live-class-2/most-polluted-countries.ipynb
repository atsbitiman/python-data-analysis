```python
import pandas as pd
```

```python
# Read Data
mp_country_1 = pd.read_csv("most-polluted-countries-1.csv")

# Tampilkan 5 data teratas
mp_country_1.head()
```

```python
# Tampilkan info
mp_country_1.info()
```

```python
# Read data
mp_country_2 = pd.read_csv("most-polluted-countries-2.csv")

# Tampilkan 5 data teratas
mp_country_2.head()
```

```python
# Tampilkan info
mp_country_2.info()
```

```python
# Buat dataframe baru untuk menyimpan hasil concat
mp_country_concat = pd.concat([mp_country_1, mp_country_2],
                              ignore_index = True)

# Menampilkan 5 data teratas
mp_country_concat.head()
```

```python
# Menampilkan 5 data terakhir
mp_country_concat.tail()
```

```python
# Buat object untuk menyimpan hasil grouping
region_group = mp_country_concat.groupby(by = 'country_region')
```

```python
# Tampilkan data petama dari setiap region
region_group.first()
```

```python
# Buat dataframe baru untuk menyimpan data region Asia
polution_asia = region_group.get_group('Asia')

# Tampilkan 5 data teratas
polution_asia.head()
```

```python
# Mendapakan data dengan polusi paling tinggi
polution_max = polution_asia[polution_asia['pollution_2023'] == polution_asia['pollution_2023'].max()]

# Print Negara dengan pollution_2023 paling tinggi
print(f"Negara dengan polusi tertinggi : {polution_max.country_name.values[0]}")

# Menampilkan data dengan pollution_2023 paling tinggi
polution_max
```

```python
# Mendapakan data dengan polusi paling rendah
polution_min = polution_asia[polution_asia['pollution_2023'] == polution_asia['pollution_2023'].min()]

# Print Negara dengan pollution_2023 paling rendah
print(f"Negara dengan polusi terendah : {polution_min.country_name.values[0]}")

# Menampilkan data dengan pollution_2023 paling rendah
polution_min
```

```python
# Mendapakan rata-rata polusi pada region Asia
polution_mean_value = polution_asia['pollution_2023'].mean()

# Mendapatkan data negara dengan polusi dibawah rata-rata
polution_mean_less = polution_asia[polution_asia['pollution_2023'] < polution_mean_value]

# Mendapatkan data negara dengan polusi diatas rata-rata
polution_mean_more = polution_asia[polution_asia['pollution_2023'] > polution_mean_value]
```

```python
# Menampilkan data negara dengan polusi dibawah rata-rata
polution_mean_less
```

```python
# Menampilkan data negara dengan polusi diatas rata-rata
polution_mean_more
```

```python
import pandas as pd
```

```python
# Read Data
yt_videos_1 = pd.read_csv("youtube-videos-1.csv")

# Tampilkan 5 data teratas
yt_videos_1.head()
```

```python
# Tampilkan info
yt_videos_1.info()
```

```python
# Read data
yt_videos_2 = pd.read_csv("youtube-videos-2.csv")

# Tampilkan 5 data teratas
yt_videos_2.head()
```

```python
# Tampilkan info
yt_videos_2.info()
```

```python
# Buat dataframe baru untuk menyimpan hasil concat
yt_videos_concat = pd.concat([yt_videos_1, yt_videos_2],
                              ignore_index = True)

# Menampilkan 5 data teratas
yt_videos_concat.head()
```

```python
# Menampilkan 5 data terakhir
yt_videos_concat.tail()
```

```python
# Merubah kolom 'published_at' menjadi datetime
yt_videos_concat['published_at'] = pd.to_datetime(yt_videos_concat['published_at'])

# Menampilkan 5 data teratas
yt_videos_concat.head()
```

```python
# Tampilkan tipe data dari data tersebut
yt_videos_concat.dtypes
```

```python
# Ekstrak data tahun dari kolom 'published_at'
yt_videos_concat["year_published"] = yt_videos_concat["published_at"].dt.year

# Ekstrak data bulan dari kolom 'published_at'
yt_videos_concat["month_published"] = yt_videos_concat["published_at"].dt.month

# Ekstrak data tanggal dari kolom 'published_at'
yt_videos_concat["day_published"] = yt_videos_concat["published_at"].dt.day

# Hapus kolom 'published_at'
yt_videos_concat.drop(columns = "published_at", inplace = True)

# Menampilkan 5 data teratas
yt_videos_concat.head()
```

```python
# Buat objek untuk menyimpan data hasil grouping
yt_videos_group = yt_videos_concat.groupby('month_published')
```

```python
# Menggunakan metode idxmax() untuk mendapatkan indeks dari baris dengan jumlah view terbanyak
idx_max = yt_videos_group.view_count.idxmax()

# Menggunakan indeks yang telah ditemukan untuk mendapatkan baris dengan jumlah view terbanyak dari DataFrame hasil concat
most_view = yt_videos_concat.loc[idx_max]

# Menampilkan baris dengan jumlah view terbanyak
most_view
```

```python
# Mendapatkan data pada bulan 2 dari hasil grouping
feb_2023_data = yt_videos_group.get_group(2)

# Urutkan data pada bulan 2 tersebut berdasarkan kolom 'view_count'
feb_2023_data_sort = feb_2023_data.sort_values(by = 'view_count',
                                               ascending = False)

# Menampilkan data
feb_2023_data_sort
```
